```python
import pandas as pd
def split_location(location):
    parts = location.split(', ')
    if len(parts) == 3:
        return parts
    elif len(parts) == 2:
        return [None] + parts
    else:
        return [None, None, parts[0]]

# Apply the split_location function to the "Location" column and expand it into separate columns
df[['City', 'Province', 'Country']] = df['location'].apply(split_location).apply(pd.Series)
# Drop the original "Location" column
df.drop(columns=['location'], inplace=True)
print(df)

# Path to save the updated combined CSV file
updated_combined_csv_path = "/content/drive/MyDrive/swm_project/jobs_new_data1.csv"
# Write the updated DataFrame to a new CSV file
df.to_csv(updated_combined_csv_path, index=False)
print("Updated combined CSV file saved successfully at:", updated_combined_csv_path)
```

```python
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk import pos_tag

def extract_skills(job_description, keywords):
    if pd.isna(job_description):  # Check if the description is NaN
        return []  # Return an empty list if the description is NaN

    # Tokenize the job description
    words = word_tokenize(str(job_description).lower())

    # Tag parts of speech
    tagged_words = pos_tag(words)

    # Extract nouns and noun phrases
    nouns = [word for word, pos in tagged_words if pos.startswith('NN')]

    # Match keywords and remove duplicates by converting to a set
    matched_skills = set([word for word in nouns if word in keywords])

    return list(matched_skills)  # Convert the set back to a list to return

# Assuming 'df1' is your DataFrame and 'description' is the column with job descriptions
df1['Skills'] = df1['description'].apply(lambda x: extract_skills(x, keywords))

# Save the DataFrame with extracted skills to a new CSV file
df1.to_csv("job_details_with_skills3.csv", index=False)

# Display the DataFrame with extracted skills
print(df1.head())
```
